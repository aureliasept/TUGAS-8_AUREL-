```python
import random
from math import comb
import matplotlib.pyplot as plt

# Fungsi untuk menghasilkan angka lotere
def generate_lottery_numbers(total_numbers, pick_count):
    return sorted(random.sample(range(1, total_numbers + 1), pick_count))

# Fungsi untuk menghitung probabilitas kemenangan
def calculate_probability(total_numbers, pick_count):
    return 1 / comb(total_numbers, pick_count)

# Simulasi sistem lotere
def lottery_simulation(total_numbers, pick_count, simulations):
    winning_numbers = generate_lottery_numbers(total_numbers, pick_count)
    matches = []

    for _ in range(simulations):
        user_numbers = generate_lottery_numbers(total_numbers, pick_count)
        match_count = len(set(winning_numbers) & set(user_numbers))
        matches.append(match_count)

    return winning_numbers, matches

# Parameter
total_numbers = 49  # Total angka yang tersedia
pick_count = 6      # Jumlah angka yang dipilih
simulations = 1000  # Jumlah simulasi

# Jalankan simulasi
winning_numbers, matches = lottery_simulation(total_numbers, pick_count, simulations)
probability = calculate_probability(total_numbers, pick_count)

# Tampilkan hasil
print(f"Angka pemenang: {winning_numbers}")
print(f"Probabilitas memenangkan lotere: 1 in {int(1/probability):,}")

# Visualisasi hasil simulasi
plt.hist(matches, bins=range(pick_count + 2), align='left', edgecolor='black')
plt.title('Distribusi Jumlah Angka yang Cocok')
plt.xlabel('Jumlah Angka Cocok')
plt.ylabel('Frekuensi')
plt.xticks(range(pick_count + 1))
plt.grid(axis='y')
plt.show()
```
